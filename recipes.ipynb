## Basic Bread

Reference: https://www.youtube.com/watch?v=ycztOGTIX-s

### Ingredients

```python
flour = c.flour(7,"cups")
water = c.water(2.4,"cups")
oil = c.oil(7,"tbsp")
sugar = c.sugar(7,"tbsp")
salt = c.salt(1,"tbsp")
yeast = c.yeast(1,"tbsp")
milk = c.milk(4,"tbsp")
```

### Make Bread

```python
bread = pd.concat([flour,water,oil,sugar,salt,yeast,milk])
bread.loc['Total']= bread.sum()
```

```python
bread
```

## Biscuits

Reference: https://sugarspunrun.com/easy-homemade-biscuits/

### Ingredients

```python
flour = c.flour(2,"cups")
sugar = c.sugar(1,"tbsp")
salt = c.salt(0.3,"tbsp")
oil = c.oil(6,"tbsp")
water = c.water(0.75,"cups")
milk = c.milk(2,"tbsp")
baking_powder = c.baking_powder(1,"tbsp")
```

### Make Biscuits

```python
biscuits = pd.concat([flour,water,oil,sugar,salt,milk,baking_powder])
biscuits.loc['Total']= biscuits.sum()
```

```python
biscuits
```

```python
biscuits / 3.8
```
