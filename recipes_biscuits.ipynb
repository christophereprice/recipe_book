```python
import convert as c
import pandas as pd
```

## Biscuits

Reference: https://sugarspunrun.com/easy-homemade-biscuits/

### Ingredients

```python
flour = c.flour(2,"cups")
sugar = c.sugar(1,"tbsp")
salt = c.salt(0.3,"tbsp")
oil = c.oil(6,"tbsp")
water = c.water(0.75,"cups")
milk = c.milk(2,"tbsp")
baking_powder = c.baking_powder(1,"tbsp")
```

### Make Biscuits

```python
biscuits = pd.concat([flour,water,oil,sugar,salt,milk,baking_powder])
biscuits.loc['Total']= biscuits.sum()
```
