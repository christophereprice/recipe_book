## Basic Bread

Reference: https://www.youtube.com/watch?v=ycztOGTIX-s

### Ingredients (./configure)

```python
flour = c.flour(3.25,"cups")
water = c.water(1.5,"cups")
oil = c.oil(1,"tbsp")
sugar = c.sugar(0.3,"tbsp")
salt = c.salt(0.5,"tbsp")
yeast = c.yeast(0.5,"tbsp")
```

### Combine Ingredients (make)

```python
ciabatta = pd.concat([flour,water,oil,sugar,salt,yeast])
ciabatta.loc['Total'] = ciabatta.sum()
```

### Deploy ciabatta

#### Set Constraints

```python
ciabatta / 2 # lets make a half recipe
```
