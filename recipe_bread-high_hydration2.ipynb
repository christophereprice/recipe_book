```python
import convert as c
import pandas as pd
```

## Basic Bread

Reference: https://www.youtube.com/watch?v=ycztOGTIX-s

### Ingredients (./configure)

```python
flour = c.flour(7,"cups")
water = c.water(3,"cups")
oil = c.oil(7,"tbsp")
sugar = c.sugar(7,"tbsp")
salt = c.salt(0.75,"tbsp")
yeast = c.yeast(2,"tbsp")
milk = c.milk(4,"tbsp")
```

### Combine Ingredients (make)

```python
bread = pd.concat([flour,water,oil,sugar,salt,yeast,milk])
bread.loc['Total'] = bread.sum()
```

### Deploy Bread

#### Set Constraints

```python
half = bread / 2 # lets make a half recipe
half
```

```python
import bakers_percentage as bp
bp.calc(half)
```
